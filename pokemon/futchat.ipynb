```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("./pokemon.csv")

print(df.head())
print(df.info())
print(df.describe())
```

```python
df
```

```python
# QUANTIDADE DE POKEMONS DE CADA TIPO

tipo_primario_counts = df['Type 1'].value_counts()
plt.figure(figsize=(12,6))
sns.barplot(x=tipo_primario_counts.index, y=tipo_primario_counts.values, palette="Set2")
plt.title("Quantidade de PokÃ©mons por Tipo PrimÃ¡rio")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("pokemon_por_tipo_primario.png")
plt.show()
```

```python
# TIPOS PRIMARIOS MAIS FORTES

forca_tipo = df.groupby("Type 1")["Attack"].mean().sort_values(ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x=forca_tipo.index, y=forca_tipo.values, palette="magma")
plt.title("Ataque MÃ©dio por Tipo PrimÃ¡rio")
plt.xticks(rotation=45)
plt.savefig("ataque_medio_tipo.png")
plt.show()
```

```python
# GRAFICO DE DISPERSAO DE ATAQUE VS DEFESA

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Attack', y='Defense', hue='Type 1')
plt.title("Ataque vs Defesa por Tipo")
plt.savefig("ataque_vs_defesa.png")
plt.show()
```

```python
# QUANTIDADE DE POKEMON LENDARIOS 

plt.figure(figsize=(6,6))
sns.countplot(x='Legendary', data=df, palette='Set1')
plt.title("PokÃ©mons LendÃ¡rios vs NÃ£o LendÃ¡rios")
plt.savefig("pokemon_lendarios.png")
plt.show()
```

```python
atributos = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
for atributo in atributos:
    top = df.loc[df[atributo].idxmax()]
    print(f"\nPokÃ©mon com maior {atributo}:")
    print(top[['Name', 'Type 1', 'Type 2', atributo]])
```

```python
atributos = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
top_pokemons = []

for atributo in atributos:
    top = df.loc[df[atributo].idxmax()]
    top_pokemons.append({
        'Atributo': atributo,
        'Nome': top['Name'],
        'Tipo 1': top['Type 1'],
        'Tipo 2': top['Type 2'],
        'Valor': top[atributo]
    })

df_top = pd.DataFrame(top_pokemons)
plt.figure(figsize=(12, 6))
sns.barplot(data=df_top, x='Atributo', y='Valor', hue='Tipo 1', dodge=False, palette='tab10')
for i, row in df_top.iterrows():
    plt.text(i, row['Valor'] + 1, row['Nome'], ha='center', fontsize=9)
plt.title("PokÃ©mons com Maior Atributo por Categoria")
plt.ylabel("Valor do Atributo")
plt.tight_layout()
plt.savefig("maior_valor_por_atributo.png")
plt.show()
```

```python
for atributo in ['Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']:
    media_tipo = df.groupby("Type 1")[atributo].mean().sort_values(ascending=False)
    plt.figure(figsize=(12,6))
    sns.barplot(x=media_tipo.index, y=media_tipo.values, palette="viridis")
    plt.title(f"{atributo} MÃ©dio por Tipo PrimÃ¡rio")
    plt.xticks(rotation=45)
    plt.savefig(f"media_{atributo.lower()}_tipo.png")
    plt.show()
```
